```python
import pandas as pd
import numpy as np
```

### Process option data

```python
### import option data and process to monthly
option_2018 = pd.read_csv('2018option.csv')
end_of_month_2018 = ['2018-01-31', '2018-02-28', '2018-03-29','2018-04-30', '2018-05-31', '2018-06-29', '2018-07-31', '2018-08-31', '2018-09-28', '2018-10-31', '2018-11-30', '2018-12-31']
option_monthly_2018 = option_2018[(option_2018['date'].isin(end_of_month_2018))]
call_monthly_2018 = pd.DataFrame(option_monthly_2018[option_monthly_2018['cp_flag'] == 'C']).reset_index(drop=True)
put_monthly_2018 = pd.DataFrame(option_monthly_2018[option_monthly_2018['cp_flag'] == 'P']).reset_index(drop=True)

option_2019 = pd.read_csv('2019option.csv')
end_of_month_2019 = ['2019-01-31', '2019-02-28', '2019-03-29','2019-04-30', '2019-05-31', '2019-06-28', '2019-07-31', '2019-08-30', '2019-09-30', '2019-10-31', '2019-11-29', '2019-12-31']
option_monthly_2019 = option_2019[(option_2019['date'].isin(end_of_month_2019))]
call_monthly_2019 = pd.DataFrame(option_monthly_2019[option_monthly_2019['cp_flag'] == 'C']).reset_index(drop=True)
put_monthly_2019 = pd.DataFrame(option_monthly_2019[option_monthly_2019['cp_flag'] == 'P']).reset_index(drop=True)

option_2020 = pd.read_csv('2020option.csv')
end_of_month_2020 = ['2020-01-31', '2020-02-28', '2020-03-31','2020-04-30', '2020-05-29', '2020-06-30', '2020-07-31', '2020-08-31', '2020-09-30', '2020-10-30', '2020-11-30', '2020-12-31']
option_monthly_2020 = option_2020[(option_2020['date'].isin(end_of_month_2020))]
call_monthly_2020 = pd.DataFrame(option_monthly_2020[option_monthly_2020['cp_flag'] == 'C']).reset_index(drop=True)
put_monthly_2020 = pd.DataFrame(option_monthly_2020[option_monthly_2020['cp_flag'] == 'P']).reset_index(drop=True)

option_2021 = pd.read_csv('2021option.csv')
end_of_month_2021 = ['2021-01-29', '2021-02-26', '2021-03-31','2021-04-30', '2021-05-28', '2021-06-30', '2021-07-30', '2021-08-31', '2021-09-30', '2021-10-29', '2021-11-30', '2021-12-31']
option_monthly_2021 = option_2021[(option_2021['date'].isin(end_of_month_2021))]
call_monthly_2021 = pd.DataFrame(option_monthly_2021[option_monthly_2021['cp_flag'] == 'C']).reset_index(drop=True)
put_monthly_2021 = pd.DataFrame(option_monthly_2021[option_monthly_2021['cp_flag'] == 'P']).reset_index(drop=True)

option_2022 = pd.read_csv('2022option.csv')
end_of_month_2022 = ['2022-01-31', '2022-02-28', '2022-03-31','2022-04-29', '2022-05-31', '2022-06-30', '2022-07-29', '2022-08-31', '2022-09-30', '2022-10-31', '2022-11-30', '2022-12-30']
option_monthly_2022 = option_2022[(option_2022['date'].isin(end_of_month_2022))]
call_monthly_2022 = pd.DataFrame(option_monthly_2022[option_monthly_2022['cp_flag'] == 'C']).reset_index(drop=True)
put_monthly_2022 = pd.DataFrame(option_monthly_2022[option_monthly_2022['cp_flag'] == 'P']).reset_index(drop=True)

call_data = pd.concat([call_monthly_2018,call_monthly_2019,call_monthly_2020,call_monthly_2021,call_monthly_2022],ignore_index=True).drop(['days','delta','cp_flag','index_flag'],axis=1)
put_data = pd.concat([put_monthly_2018,put_monthly_2019,put_monthly_2020,put_monthly_2021,put_monthly_2022],ignore_index=True).drop(['days','delta','cp_flag','index_flag'],axis=1)
call_data.dropna(axis=0,inplace=True)
put_data.dropna(axis=0,inplace=True)
```

```python
### There are some rows without effective ticker. Drop them.
option_data = pd.merge(call_data, put_data, how='left',on=['ticker','date'] )
option_data.drop(['secid_y','cusip_y'],axis=1,inplace=True)
option_data.rename(columns={
    'secid_x': 'secid',
    'impl_volatility_x':'CVOL',
    'cusip_x':'cusip',
    'impl_volatility_y':'PVOL'
}, inplace=True)
option_data.reset_index(drop=True, inplace=True)
```

### Process alpha data

```python
### import CAPM ALPHA data
capm_alpha = pd.read_csv('CAPM_alpha.csv')
capm_alpha.drop(['PERMNO','n','ivol','tvol','R2','exret'], axis=1, inplace=True)
capm_alpha.dropna(axis=0,inplace=True)
```

```python
### import ff3 ALPHA data
ff3_alpha = pd.read_csv('ff3_alpha.csv')
ff3_alpha.drop(['PERMNO','n','ivol','tvol','R2','exret'], axis=1, inplace=True)
ff3_alpha.dropna(axis=0,inplace=True)
```

```python
alpha_data = pd.merge(capm_alpha, ff3_alpha, how='left', on=['TICKER','DATE'])
alpha_data.drop(['RET_y'],axis=1,inplace=True)
alpha_data.rename(columns={
    'DATE':'date',
    'RET_x':'return',
    'b_mkt_x':'beta_market',
    'TICKER':'ticker',
    'alpha_x':'capm_ahpha',
    'alpha_y':'ff3_alpha',
    'b_mkt_y':'beta_market_ff3',
    'b_smb':'beta_smb',
    'b_hml':'beta_hml'
},inplace=True)
```

### Merge option data and alpha data

```python
data = pd.merge(option_data, alpha_data, how='left', on=['ticker','date'])
data.dropna(axis=0,inplace=True)
data.drop_duplicates(subset=['date','ticker'], inplace=True)
data.reset_index(drop=True, inplace=True)
```

### compute delta CVOL and delta PVOL

```python
data['date'] = pd.to_datetime(data['date'])
all_dates = data['date'].sort_values().unique().tolist()
data['delta_CVOL']=np.NaN
data['delta_PVOL']=np.NaN
data['future_return']=np.NaN
```

```python
### track the initil date that an equity occurs. Dates before initial dates do not have delta_CVOL and delta_PVOL
initial_dates = data.groupby('ticker')['date'].min()

end_dates = data.groupby('ticker')['date'].max()
```

```python
tickers = data['ticker'].unique()    #### 6491 tickers

#### We them divide data into 8 groups to calculate delta_CVOL and delta_PVOL. The sample is too big so that ,y computer can not run them all at one time.
```

```python
### 1 group to compute delta_CVOL and delta_PVOL

group1 = tickers[0:800]
data1 = data[data['ticker'].isin(group1)]
for index,row in data1.iterrows():
    if all_dates.index(row['date']) != all_dates.index(initial_dates[row['ticker']]) :  ### can not calculate delta_CVOL for initial date
        try: 
            data1.at[index, 'delta_CVOL'] = row['CVOL']-data1[(data1['ticker']==row['ticker'])&(data1['date']==all_dates[(all_dates.index(row['date']))-1])]['CVOL'].tolist()[0]
            data1.at[index, 'delta_PVOL'] = row['PVOL']-data1[(data1['ticker']==row['ticker'])&(data1['date']==all_dates[(all_dates.index(row['date']))-1])]['PVOL'].tolist()[0]
        except:
            print('error occured due to lack of data')
    if all_dates.index(row['date']) != all_dates.index(end_dates[row['ticker']]) :      ### can not calculate future_return for end date
        try:
            data1.at[index, 'future_return'] = data1[(data1['ticker']==row['ticker'])&(data1['date']==all_dates[(all_dates.index(row['date']))+1])]['return'].tolist()[0] 
        except:
            print('have no next-month data')
        
```
