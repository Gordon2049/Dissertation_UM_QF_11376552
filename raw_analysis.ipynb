```python
### 2022
data_all = pd.read_csv('2022option.csv')
data_all.head()
end_of_month = ['2022-01-31', '2022-02-28', '2022-03-31','2022-04-29', '2022-05-31', '2022-06-30', '2022-07-29', '2022-08-31', '2022-09-30', '2022-10-31', '2022-11-30', '2022-12-30']
```

```python
monthly_data = data_all[(data_all['date'].isin(end_of_month))]
call_data = pd.DataFrame(monthly_data[monthly_data['cp_flag'] == 'C']).reset_index(drop=True)
put_data = pd.DataFrame(monthly_data[monthly_data['cp_flag'] == 'P']).reset_index(drop=True)
```

```python
## number of stocks
len(data_all['secid'].unique())
```

```python
## compute call option data
call_average = call_data['impl_volatility'].mean()
call_sd = call_data['impl_volatility'].std()
pd.DataFrame({
    'CVOL Avarage':[call_average],
    'CVOL stdev':[call_sd]
})
```

```python
## compute put option data
put_average = put_data['impl_volatility'].mean()
put_sd = put_data['impl_volatility'].std()
pd.DataFrame({
    'PVOL Avarage':[put_average],
    'PVOL stdev':[put_sd]
})
```

```python
## compute IV spread data
IVS = pd.DataFrame({
    'Ticker':call_data['ticker'],
    'impl_volatility_spread':call_data['impl_volatility']-put_data['impl_volatility']
})
IVS_average = IVS['impl_volatility_spread'].mean()
IVS_sd = IVS['impl_volatility_spread'].std()
pd.DataFrame({
    'IVS Avarage':[IVS_average],
    'IVS stdev':[IVS_sd]
})
```
